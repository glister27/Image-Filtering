## Image Filtering Using Average Filter and MedianFilter

These Filters are basically used to improve the quality of the image and reduces noise in the image.Reducing noise in an image is a preprocessing technique in image processing.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading thr image using matplotlib.pyploy
img=plt.imread("minion.jpeg")
plt.imshow(img)
```

If we apply our filters on the three channels of the image it will be complex so we write a fuction that will take the three channels and merge them into one channel by taking the average of the three channels for each pixel

```python
def  bw(img):
    I=np.zeros((img.shape[0],img.shape[1]))
    for i in range (img.shape[0]):
        for j in range (img.shape[1]):
            I[i][j]=np.round(np.mean(img[i][j]))
            
    return I
```

```python
i=bw(img)
plt.imshow(i)
i.shape
```

We have gotten our one channel image so we can write our functions now

## Average Flilter

The average filter takes a pixel and represent it by the average of its surrounding  3 X 3 sorrounding pixels but the pixels on the borders are not taken into consideration because they dont have 3 X 3 surroubding neighbors.
In our case we just ignore them thereby reducing the size of image.

```python
#filtre carre avec la moyyenne
def moyhexa(img):
    L=np.zeros(img.shape)
  
    for i in range(1,img.shape[0]-1,1):
        for j in range(1,img.shape[1]-2,1):
            L[i][j]=np.mean((img[i-1][j-1],img[i-1][j+1],img[i][j-2],img[i][j],img[i][j+2],img[i+1][j-1],img[i+1][j+1]))
    return L
```

```python
c=moyhexa(i)
plt.imshow(c)
```

## Median Filter

The Median Filter takes a pixel and represent it by the median of its surrounding  3 X 3 sorrounding pixels but the pixels on the borders are not taken into consideration because they dont have 3 X 3 surroubding neighbors.
In our case we just ignore them thereby reducing the size of image.

```python
#filtre hexa en utilisant numpy avec median
def medhex(img):
    L=np.zeros(img.shape)
    m=[]
    p=img.shape[0]-1
    q=img.shape[1]-2
    L[0,:]=img[0,:]
    L[:,0]=img[:,0]
    L[p-1,:]=img[p-1,:]
    L[:,p-1]=img[:,p-1]
    for i in range(1,p-1,1):
        for j in range(2,q-1,1):
            L[i][j]=np.median((img[i-1][j-1],img[i-1][j+1],img[i][j-2],img[i][j],img[i][j+2],img[i+1][j-1],img[i+1][j+1]))          
                
    return L
```
